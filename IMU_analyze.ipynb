```python
import json
import matplotlib.pyplot as plt
```

```python
# data = CONFIDENTIAL_SAMPLE_DATA
# file_path = 'Data/walking/IMU_000_CONFIDENTIAL.json'
```

```python


def extract_and_plot(json_string):
    try:
        data = json.loads(json_string)
    except json.JSONDecodeError:
        return "Invalid JSON format"

    # Initialize lists to store the extracted data
    times = []
    x_values = []
    y_values = []
    z_values = []

    # Extract data from each record
    for record in data:
        if all(key in record for key in ['t', 'i', 'v']):
            # Assuming the first value in 'i' is the time
            times.append(record['i'][0])
            x_values.append(record['v'][0])
            y_values.append(record['v'][1])
            z_values.append(record['v'][2])

    # Plotting
    plt.figure(figsize=(20, 6))
    plt.plot(times, x_values, label='X-axis')
    plt.plot(times, y_values, label='Y-axis')
    plt.plot(times, z_values, label='Z-axis')

    plt.xlabel('Time (ms)')
    plt.ylabel('Values')
    plt.title('X, Y, Z values over Time')
    plt.legend()
    plt.show()


# Extract data and plot
extract_and_plot(json_string)
```
